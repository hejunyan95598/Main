```python
x = [1,2,3]
y = x # This does not create a new list, but rather assign y as a reference of x, both point to the same list
y[0] = 100
x # x also changes

# The following line can use explicit copy to solve the issue
x = [1,2,3]
y = list(x)
y [0] = 100
x # x does not change
```
